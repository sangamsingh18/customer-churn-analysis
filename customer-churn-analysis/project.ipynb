```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

df=pd.read_csv('customer churn.csv')
print(df)
print(df.head(20))
df.info()
```

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"
    
    df["SeniorCitizen"]=["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("total number of count by churn")
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("percentage of churned customeres")
plt.show()
```

```python
plt.figure(figsize = (2,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "SeniorCitizen", data = df)
plt.title("Churn by senior centizen ")
plt.show()
```

```python
plt.figure(figsize=(2,2))

# value_counts normalize karke percentage nikale
percent_data = df["SeniorCitizen"].value_counts(normalize=True) * 100

# barplot banao
sns.barplot(x=percent_data.index, y=percent_data.values)

plt.ylabel("Percentage (%)")
plt.xlabel("SeniorCitizen")
plt.title("Churn by Senior Citizen (Percentage)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# 3 row x 3 column ka grid
fig, axes = plt.subplots(3, 3, figsize=(20,20))
axes = axes.flatten()   # 2D array ko 1D banane ke liye

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i],hue='Churn')
    axes[i].set_title(f"{col}")
    axes[i].tick_params(axis='x', rotation=30)  # x labels ko tilt

plt.tight_layout()
plt.show()
```

Fiber optic Internet users me churn sabse zyada hai, DSL users relatively stable hain.

Online Security, Backup, Device Protection aur TechSupport hone par churn kam hota hai.

Jinke paas ye services nahi hoti, unme churn kaafi high hai.

Phone service aur multiple lines wale customers bhi churn karte hain, lekin impact moderate hai.

StreamingTV/Movies ka churn pe zyada effect nahi hai, dono me churn similar hai.
